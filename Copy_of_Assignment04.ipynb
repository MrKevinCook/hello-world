```
df = pand.read_csv('https://raw.githubusercontent.com/frankData612/pandas_for_everyone/master/data/tesla_stock_yahoo.csv')
```

```
hist, ax = plt.subplots()

ax = sns.histplot(df['Volume'],kde=False)

ax.set_title('Volume and Price')

ax.set_xlabel('Volume in Millions')

ax.set_ylabel('High Price')

plt.show()

#Histogram that shows the high price and volume.
#That is, the correlation between volume and price movement.
```

```
dis, ax = plt.subplots()

ax = sns.distplot(df['Volume'], hist=False)

ax.set_title('Total Volume Density')

ax.set_xlabel('Volume in Millions')

ax.set_ylabel('Unit Probability')

plt.show()

#Denisty Plot to simply smooth out overlapping plots so the area under the curve is 1.
```

```
fig = sns.pairplot(df)

#Pairwise Relationships to plot a scatterplot between each variable.
```

```
pair_grid = sns.PairGrid(df)

pair_grid = pair_grid.map_upper(sns.regplot)

pair_grid = pair_grid.map_lower(sns.kdeplot)

pair_grid = pair_grid.map_diag(sns.distplot, rug=True)

plt.show()
```
